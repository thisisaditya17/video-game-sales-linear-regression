```python
import pandas as pd
```

```python
game_sales = pd.read_csv('vgchartz-2024.csv')
```

```python
game_sales
```

```python
game_sales = game_sales[["title", "console", "genre", "publisher", "developer", "total_sales", "na_sales", "jp_sales","pal_sales", "other_sales", "release_date"]]
```

```python
numeric_game_sales = game_sales.select_dtypes(include=['float64', 'int64'])
numeric_game_sales.corr()["total_sales"]
```

```python
import seaborn as sns
```

```python
sns.lmplot( x="na_sales", y="total_sales", data=game_sales, fit_reg=True, ci=None)
```

```python
game_sales[game_sales.isnull().any(axis=1)]
```

```python
game_sales = game_sales.dropna()
```

```python
game_sales

game_sales[game_sales["release_date"] > "2010-01-01"]
```

```python
training_set = game_sales[game_sales["release_date"] < "2017-01-01"].copy()
testing_set = game_sales[game_sales["release_date"] >= "2017-01-01"].copy()

training_set.shape
```

```python
testing_set.shape
```

```python
from sklearn.linear_model import LinearRegression
regress = LinearRegression()
```

```python
predictors = ["na_sales", "other_sales", "pal_sales"]
target = "total_sales"
```

```python
regress.fit(training_set[predictors], training_set[target])
```

```python
predictions = regress.predict(testing_set[predictors])

testing_set["predicted_total_sales"] = predictions
testing_set
```

```python
testing_set.loc[testing_set["predicted_total_sales"] < 0, "predicted_total_sales"] = 0
testing_set["predicted_total_sales"].round(2)
```

```python
testing_set
```

```python
from sklearn.metrics import mean_absolute_error as mae

error = mae(testing_set[target], testing_set["predicted_total_sales"])
testing_set["predicted_total_sales"].round(2)
error
```

```python
game_sales.describe()["na_sales"]
```

```python
absolute_error = (testing_set[target] - testing_set["predicted_total_sales"].round(2)).abs()
absolute_error
```

```python
error_by_publisher = absolute_error.groupby(testing_set["publisher"]).mean()
error_by_publisher
```

```python
total_sales_by_publisher = testing_set["total_sales"].groupby(testing_set["publisher"]).mean()
error_ratio = error_by_publisher / total_sales_by_publisher
error_ratio
```

```python
import numpy as np
error_ratio = error_ratio[~pd.isnull(error_ratio)]
error_ratio = error_ratio[np.isfinite(error_ratio)]
error_ratio
```

```python
error_ratio.plot.hist()
```

```python
error_ratio.sort_values(ascending=False)
```
